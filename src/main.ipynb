```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from ConvexHull import *
from MainProgram import *

#inisialisasi variabel global
convex_hull = ConvexHull()
```

```python
# Test 1
print("Dataset Pertama : ")
data1 = datasets.load_wine()
# Create a DataFrame
df1 = pd.DataFrame(data1.data, columns=data1.feature_names)
df1['Target'] = pd.DataFrame(data1.target)
print(df1.shape)
df1.head()

# Visualisasi hasil ConvexHull
plt.figure(figsize = (10, 6))
colors = ['b','r','g']
plt.title('malic_acid vs alcohol')
plt.xlabel(data1.feature_names[0])
plt.ylabel(data1.feature_names[1])
for i in range(len(data1.target_names)):
    bucket1 = df1[df1['Target'] == i]
    print(bucket1)
    bucket1 = bucket1.iloc[:, [0,1]].values
    plt.plot(bucket1[:,0], bucket1[:,1], 'o')
    convexhull1 = convex_hull(bucket1)
    print()
    print ("Vertex dari convex hull adalah : ")
    print (convexhull1)
    print()
    # Close the hull during plot by making the head and tail as the same point
    convexhull1 = np.vstack([convexhull1, convexhull1[0]])
    plt.plot(convexhull1[:,0], convexhull1[:,1], colors[i])
plt.show()
```
